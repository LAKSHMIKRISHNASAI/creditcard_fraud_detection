```python
#label-1: fraudlent transactions
#label-0: Normal Transactions
#separating the both of data
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
#now to  build sample new dataset of normal transactions and fraud transaction
legit_sample= legit.sample(n=492)
```

```python
new_dataset= pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
x=new_dataset.drop(columns='Class',axis=1)
#axis=0 represents rows
#axis=1 represents columns
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
x_train,x_test,y_train,y_test= train_test_split(x,y, random_state=2,test_size=0.2,stratify=y )
```

```python
#80% data store in x train
#20% data store in y train
print(x_train.shape,x_test.shape)
```

```python
#logistic regression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print("accuracy:",training_data_accuracy)
```

```python
 #accuracy on test data
x_test_predict=model.predict(x_test)
accuracy_testing_data=accuracy_score(x_test_predict,y_test)
```

```python
print("accuracy of y:",accuracy_testing_data)
```
